```python
import pylab
from bregman.suite import *
import os
import wave
from pocketsphinx import AudioFile, get_model_path, get_data_path
from pocketsphinx.pocketsphinx import *
from sphinxbase.sphinxbase import *
from pydub import AudioSegment
from pydub.silence import split_on_silence
import IPython.display as ipd
```

```python
def match_target_amplitude(aChunk, target_dBFS):
    ''' Normalize given audio chunk '''
    change_in_dBFS = target_dBFS - aChunk.dBFS
    return aChunk.apply_gain(change_in_dBFS)

filename = '/Users/aaronkarp/Desktop/3.wav'
min_len = 80
thresh = -35
max_len = 1.8

song = AudioSegment.from_wav(filename)
song = match_target_amplitude(song, -20.0)

#split track where silence is 2 seconds or more and get chunks

chunks = split_on_silence(song,
    min_silence_len=min_len,
    silence_thresh=thresh
)

print("Split into " + str(len(chunks)) + " chunks")

#Process each chunk per requirements
removed = 0
removed_time = 0
total_time = 0
i = 0
for k,chunk in enumerate(chunks):
    i += 1
    time = len(chunk)/1000.0
    if time > max_len:
        removed_time += time
        removed += 1
        i -= 1
        continue

    #Export audio chunk with new bitrate
    total_time += time
    print("exporting " + newname.format(i) + "  :  " + str(time) + "s")
    newname = filename[:-4] + '_{0}' + filename[-4:]
    chunk.export(newname.format(i), format="wav")
print
print("removed " + str(removed) + " chunks")
print("removed " + str(removed_time) + "s of audio")
print("total time exported: " + str(total_time) + "s")
```
