```python
## WRITE STREAM DATA TO FILE

num_files_in_stream = 50
NUM_CLASSES_OVERRIDE = 2  # If 0, use ALL classes; else, use first n classes only
num_used = [0, 0]

start = time.time()
total_time = time.time()

def generate_used_stream(used_files, classes_to_use):
    to_use = randrange(NUM_CLASSES_OVERRIDE)
    while num_used[to_use] > num_files_in_stream:
        to_use = randrange(NUM_CLASSES_OVERRIDE)
    num_used[to_use] += 1
    random_index = classes_to_use[to_use]
    matched_class = to_use
    to_check = [used_files[random_index], str(matched_class), used_files[random_index]]
    return to_check

def generate_data_file(name, classes_to_use=None):

    # Load used sounds
    used_f = open(filedir + '/sampledFiles.txt', "r")
    used_files = used_f.read().split('\n')
    #used_fs = ["../testSounds/nonspeech/bigben.wav", "../testSounds/nonspeech/amen.wav"]

    open(filedir + name, 'w').close()  # Delete previous info

    with open(filedir + name, "a") as training_file:
        for i in range(num_files_in_stream):
            info = generate_used_stream(used_fs, classes_to_use)
            to_write = info[0] + " -_- " + info[1] + " -_- " + info[2] + " \n"
            training_file.write(to_write)

f_o = open(filedir + '/sampledFiles.txt', "r")
used_fs = f_o.read().split('\n')
classes_to_use = []
for i in range(NUM_CLASSES_OVERRIDE):
    classes_to_use.append(random.randint(0, len(used_fs)))  # THIS WILL NOT WORK IF WE USED MATCHED NON-PERFECT PAIRS
#used_fs = ["../testSounds/nonspeech/bigben.wav", "../testSounds/nonspeech/amen.wav"]
print("Using classes: ")
print(classes_to_use)
for c in classes_to_use:
    y1,sr1 = librosa.load(used_fs[c])
    print(c)
    ipd.display(ipd.Audio(y1, rate = sr1)) # load query file

generate_data_file("trainingInfo.txt", classes_to_use)
generate_data_file("testingInfo.txt", classes_to_use)

print(num_used)

# code you want to evaluate
end = time.time()
print("Time elapsed: " + str(end-start) + " seconds, or " + str((end-start)/60.0) + " minutes")
```

```python
## READ STREAM DATA FROM FILE

def load_stream(name, filedir='../'):
    training_file = open(filedir + name, "r")
    info = training_file.read().split('\n')
    data_matrix = []
    for inf in info:
        segs = inf.split('-_-')
        if len(segs) > 1:
            try:
                data_matrix.append([segs[0].strip(), segs[1].strip(), segs[2].strip()])
            except Exception as e:
                #print(e)
                print(inf)
        else:
            print("\nBlank Segment at File Start")
    training_file.close()
    return data_matrix

def generate_composite_stream(audio_matrix):
    cur_percentage = 0
    data_dict = {}
    composite_signal = np.zeros(1)
    composite_matches = np.full(composite_signal.shape, 0)
    
    for i in range(len(audio_matrix)):
        if (round(i / len(audio_matrix) * 100) != cur_percentage):
            cur_percentage = round(i / len(audio_matrix) * 100)
            print(str(cur_percentage) + "%     ", end='')
            
        filename = audio_matrix[i][0]
        if filename in data_dict:
            y = data_dict.get(filename)
        else:
            y,sr = librosa.load(filename)
            data_dict[filename] = y
        composite_signal = np.concatenate((composite_signal, y))
        composite_matches = np.concatenate((composite_matches, np.full(y.shape, audio_matrix[i][1])))
    return composite_signal, composite_matches
    
start = time.time()

audio_matrix = load_stream(name="trainingInfo.txt")
composite_signal, composite_matches = generate_composite_stream(audio_matrix)
audio_matrix_test = load_stream(name="testingInfo.txt")
composite_signal_test, composite_matches_test = generate_composite_stream(audio_matrix_test)

composite_matches 

end = time.time()
print("\nTime elapsed: " + str(end-start) + " seconds, or " + str((end-start)/60.0) + " minutes")
```

```python
start = time.time()
n_mels = 128


def batch(signal, matches):
    # Generate batched data list for signal/matches
    
    data = []
    classes = []
    spec = get_spectrogram(signal, 22050, n_mels=n_mels)
    cur_col = 0
    transposed = spec.T
    scaler = MinMaxScaler(feature_range=(0, 1))
    transposed = scaler.fit_transform(transposed)
    batch_size = 10
    cur_percentage = 0
    for col in transposed:
        if cur_col + batch_size >= len(transposed):
            break
        if (round(cur_col / len(transposed) * 100) != cur_percentage):
            cur_percentage = round(cur_col / len(transposed) * 100)
            print(str(cur_percentage) + "%     ", end='')
        comp_cols = [col]
        for i in range(1,batch_size):
            comp_cols.append(transposed[cur_col + i])
        data.append(comp_cols)
        classes.append(int(matches[int(cur_col / len(transposed) * len(matches))]))
        cur_col += 1
    return data, classes

print("Training Data")
training_data, training_classes = batch(composite_signal, composite_matches)
print("\nTesting Data")
testing_data, testing_classes = batch(composite_signal_test, composite_matches_test)
print()
max_spec_length = len(training_data[0])
print()
training_data = np.array(training_data)
testing_data = np.array(testing_data)

end = time.time()
print("Time elapsed: " + str(end-start) + " seconds, or " + str((end-start)/60.0) + " minutes")
```

```python
# Make/test the LSTM

lstm_out = 50
batch_size = 128
embedding_vector_length = 32
dropout = 0.2
dropout_r = 0.2

num_classes = NUM_CLASSES_OVERRIDE
number_inputs = NUM_CLASSES_OVERRIDE

print(str(number_inputs) + " " + str(embedding_vector_length) + " " + str(len(training_data[0])) + " " + str(dropout) + 
      " " + str(lstm_out) + " " + str(dropout_r) + " " + str(num_classes))
    
model = Sequential()
#model.add(Embedding(number_inputs, embedding_vector_length, input_length = max_spec_length))
model.add(LSTM(lstm_out, input_shape=(10, n_mels), dropout = dropout, recurrent_dropout = dropout_r))
model.add(Dense(num_classes-1,activation='sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer='adam',metrics = ['accuracy'])
print(model.summary())

model.fit(training_data, training_classes, epochs=3, batch_size=batch_size)
scores = model.evaluate(testing_data, testing_classes)
print("Accuracy: %.2f%%" % (scores[1]*100))

total_end = time.time()
print("Total Time elapsed: " + str((total_end-total_time)/60.0) + " minutes")
```
